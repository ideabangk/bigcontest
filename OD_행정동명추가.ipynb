```python
import pandas as pd

df= pd.read_csv('데이터분석 분야_로우데이터/od_20230901_10/od_20230901_1.csv')
df
```

```python
# 행정동코드
df_dong = pd.read_excel('행정동코드/KIKcd_H.20240118(말소코드포함).xlsx')
df_dong = df_dong.loc[:,['행정동코드','읍면동명']]

# 행정동코드에서 시도/시군구 단위 제거
df_dong_dropna = df_dong.dropna().reset_index(drop=True)
df_dong_dropna

# 테이블 조인 위해 컬럼명 변경
df_dong.rename(columns={'행정동코드': 'hdong_cd', '읍면동명': 'hdong_nm'}, inplace=True)

# 기존 테이블에 출발지 행정동명 추가
df_merged = df.merge(
    df_dong.rename(columns={'hdong_cd': 'origin_hdong_cd', 'hdong_nm': 'origin_hdong_nm'}),
    on='origin_hdong_cd',
    how='left'
)

# 기존 테이블에 도착지 행정동명 추가
df_merged = df_merged.merge(
    df_dong.rename(columns={'hdong_cd': 'dest_hdong_cd', 'hdong_nm': 'dest_hdong_nm'}),
    on='dest_hdong_cd',
    how='left'
)

# 출발지와 도착지 행정동명을 결합한 새로운 컬럼 생성
df_merged['OD_Pair'] = df_merged['origin_hdong_nm'] + ' -> ' + df_merged['dest_hdong_nm']
```

```python
df_merged
df_merged.to_csv('od_행정동명추가_0901.csv',index=False)
```
